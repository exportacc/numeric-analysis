```python
import numpy as np
import matplotlib.pyplot as pt
from matplotlib import style
style.use('seaborn')

f = lambda t , y : t * y + t**3

to , te = 0 , 1 
n = 10
dt = (te - to) / n

w1 , w2 , w3 = np.zeros(n+1) , np.zeros(n+1) ,np.zeros(n+1)
w1[0] , w2[0] , w3[0] = 1 , 1 , 1 # init 

t = np.linspace(to,te,n+1)
print("{:s} \t {:8s} \t {:8s} \t {:8s}".format('t','euler','trapezoid','midpoint'))
print('-'*50)

for i in range(n):
    w1[i+1] = w1[i] + dt * f(t[i],w1[i])
    w2[i+1] = w2[i] + 0.5 * dt * ( f(t[i],w2[i]) + f( t[i]+dt,w2[i]+dt*f(t[i],w2[i]) ) )
    w3[i+1] = w3[i] + dt * f(t[i]+0.5*dt , w3[i] + 0.5*dt*f(t[i],w3[i]))
    print("{:.2f} \t {:.8f} \t {:.8f} \t {:8f}".format(t[i] ,w1[i] , w2[i] , w3[i]))
print("{:.2f} \t {:.8f} \t {:.8f} \t {:8f}".format(t[n] ,w1[n] , w2[n] , w3[n]))    
```

```python
pt.figure(figsize=(15,5))
pt.plot(t,w1,'r--',label = 'euler')
pt.plot(t,w2,'b-',label = 'trapezoid')
pt.plot(t,w3,'y^--',label = 'midpoint')
pt.legend()
pt.show()
```
