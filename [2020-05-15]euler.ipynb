尤拉法最主要的差別在于 : 

$$ w_{i+1} = w_i + f(t_i) * dt $$ 主要是求函數逼近值

$$ w_{i+1} = w_i + f(w_i) * dt $$ 主要求誤差 想要穩定的狀態

```python
import numpy as np
import matplotlib.pyplot as pt
from matplotlib import style
style.use('seaborn')

f = lambda x : 10*(1-x) # 函數 y 的微分

for n in np.array([10,100,1000,10000,100000,1000000]): # 當 n = 10 , 100 , 1000 , 10000 , 100000 , 1000000 則逼近值等於多少 ? 
    t0 , tn  = 0 , 1 # 從 0 開始 逼近 到  1 
    dt = (tn - t0) / n # 從 0 開始 逼近到 1 分成 n 等分
    t = np.linspace(t0,tn,n+1) # t 軸的分佈情況
    w = np.zeros(n+1) # 把全部值初始化放 0 

    w[0] = 0   # 預設值 y(0) = 1 
    for i in range(n): # for loop 逼近值
        w[i+1] = w[i] + f(t[i]) * dt
    print('n = {} , w = {}'.format(n,w[-1]))
    pt.plot(t,w)
y = lambda x : 10 * (x - x**2 / 2) # 假設知道真正的函數 y
pt.plot(t,y(t),'k--') # 用y函數的值 畫出真實的值 去比較 euler thoery 得出的值 是否相等
pt.show() # 可加 可不加 , 但下次的圖形不會初始化 會叠加
```
